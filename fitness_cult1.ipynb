```python
import pandas as pd
df_april_may = pd.read_csv("Classes April-May 2018.csv")
df_april_may.head()
```

```python
df_april_may.columns = [
    "ActivitySiteID", 
    "ActivityDescription", 
    "BookingEndDate", 
    "BookingStartTime", 
    "MaxBookees", 
    "NumberBooked", 
    "PriceINR"
]
df_april_may.head()
```

```python
df_april_may["BookingEndDate"] = pd.to_datetime(df_april_may["BookingEndDate"], errors='coerce')
df_april_may["BookingStartTime"] = pd.to_datetime(df_april_may["BookingStartTime"], format='%H:%M:%S', errors='coerce').dt.time
df_april_may.head()
```

```python
df_april_may["PriceINR"].fillna(df_april_may["PriceINR"].median(), inplace=True)
df_april_may.head()
df_april_may["DayOfWeek"] = df_april_may["BookingEndDate"].dt.day_name()
df_april_may["OccupancyRate"] = df_april_may["NumberBooked"] / df_april_may["MaxBookees"]
df_april_may.head()
```

```python
df_april_may.to_csv("Classes_April_to_May_2018.csv", index=False)
```

```python
import pandas as pd
df_june = pd.read_csv("Classes June 2018.csv")
df_june.head()
```

```python
df_june.columns = [
    "ActivitySiteID", 
    "ActivityDescription", 
    "BookingEndDate", 
    "BookingStartTime", 
    "MaxBookees", 
    "NumberBooked", 
    "PriceINR"
]
df_june.head()
```

```python
df_june["BookingEndDate"] = pd.to_datetime(df_june["BookingEndDate"], errors='coerce')
df_june["BookingStartTime"] = pd.to_datetime(df_june["BookingStartTime"], format="%H:%M:%S", errors='coerce').dt.time
df_june.head()
```

```python
df_june["DayOfWeek"] = df_june["BookingEndDate"].dt.day_name()
df_june["OccupancyRate"] = df_june["NumberBooked"] / df_june["MaxBookees"]
df_june.head()
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
import matplotlib.pyplot as plt
df_model = df_combined[(df_combined["PriceINR"] > 0) & (df_combined["NumberBooked"] > 0)].copy()
df_model["log_price"] = np.log(df_model["PriceINR"])
df_model["log_demand"] = np.log(df_model["NumberBooked"])
X = df_model[["log_price"]]
y = df_model["log_demand"]
```

```python
model = LinearRegression()
model.fit(X, y)

elasticity = model.coef_[0]
intercept = model.intercept_

print(f"Estimated Price Elasticity of Demand: {elasticity:.4f}")

```

```python
y_pred = model.predict(X)
residuals = y - y_pred
r2 = r2_score(y, y_pred)
mse = mean_squared_error(y, y_pred)

print(f"R²: {r2:.4f}")
print(f"Mean Squared Error: {mse:.4f}")

plt.figure(figsize=(8, 6))
plt.scatter(y_pred, residuals, alpha=0.5)
plt.axhline(0, color='red', linestyle='--')
plt.title("Residual Plot")
plt.xlabel("Predicted Log(Demand)")
plt.ylabel("Residuals")
plt.tight_layout()
plt.show()
```

```python
df_timeseries = df_combined.groupby("BookingEndDate")["NumberBooked"].sum().reset_index()

df_timeseries = df_timeseries.set_index("BookingEndDate")
df_timeseries = df_timeseries.asfreq("D")  
df_timeseries = df_timeseries.fillna(0)   

df_timeseries.head()
```

```python
from prophet import Prophet
df_prophet = df_timeseries.reset_index().rename(columns={"BookingEndDate": "ds", "NumberBooked": "y"})
model = Prophet(daily_seasonality=True)
model.fit(df_prophet)

future = model.make_future_dataframe(periods=15) 
forecast = model.predict(future)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
df_merge = pd.merge(df_prophet, forecast[['ds', 'yhat']], on='ds')
mse = mean_squared_error(df_merge['y'], df_merge['yhat'])
mae = mean_absolute_error(df_merge['y'], df_merge['yhat'])

print(f"Mean Squared Error: {mse:.2f}")
print(f"Mean Absolute Error: {mae:.2f}")
```

```python
model.plot(forecast)
plt.title("Demand Forecast for Fitness Classes")
plt.xlabel("Date")
plt.ylabel("Number of Bookings")
plt.tight_layout()
plt.show()

model.plot_components(forecast)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
def dynamic_price(row, base_price, elasticity, forecasted_demand, threshold_high=0.75, threshold_low=0.25):
    """
    Adjust price based on forecasted demand and price elasticity.
    
    - If demand > threshold_high, increase price.
    - If demand < threshold_low, decrease price.
    """
    adjustment_factor = 1.0
    if row['OccupancyRate'] > threshold_high:
        adjustment_factor += 0.1 * abs(elasticity)
    elif row['OccupancyRate'] < threshold_low:
        adjustment_factor -= 0.1 * abs(elasticity)
    predicted_demand = forecasted_demand.get(str(row["BookingEndDate"].date()), 1)
    if predicted_demand > df_prophet["y"].mean():
        adjustment_factor += 0.05
    elif predicted_demand < df_prophet["y"].mean():
        adjustment_factor -= 0.05

    return max(base_price * adjustment_factor, 50)
forecast_demand_map = dict(zip(forecast['ds'].dt.date.astype(str), forecast['yhat']))
df_combined["DynamicPrice"] = df_combined.apply(
    lambda row: dynamic_price(row, row["PriceINR"], elasticity, forecast_demand_map),
    axis=1
)
plt.figure(figsize=(10, 6))
sns.histplot(df_combined["PriceINR"], color='blue', label='Original Price', kde=True)
sns.histplot(df_combined["DynamicPrice"], color='orange', label='Dynamic Price', kde=True)
plt.legend()
plt.title("Comparison of Original vs Dynamic Prices")
plt.xlabel("Price (INR)")
plt.tight_layout()
plt.show()
```
