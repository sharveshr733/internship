```python
import pandas as pd
df_april_may = pd.read_csv("Classes April-May 2018.csv")
df_april_may.head()
```

```python
df_april_may.columns = [
    "ActivitySiteID", 
    "ActivityDescription", 
    "BookingEndDate", 
    "BookingStartTime", 
    "MaxBookees", 
    "NumberBooked", 
    "PriceINR"
]
df_april_may.head()
```

```python
df_april_may["BookingEndDate"] = pd.to_datetime(df_april_may["BookingEndDate"], errors='coerce')
df_april_may["BookingStartTime"] = pd.to_datetime(df_april_may["BookingStartTime"], format='%H:%M:%S', errors='coerce').dt.time
df_april_may.head()
```

```python
df_april_may["PriceINR"].fillna(df_april_may["PriceINR"].median(), inplace=True)
df_april_may.head()
df_april_may["DayOfWeek"] = df_april_may["BookingEndDate"].dt.day_name()
df_april_may["OccupancyRate"] = df_april_may["NumberBooked"] / df_april_may["MaxBookees"]
df_april_may.head()
```

```python
df_april_may.to_csv("Classes_April_to_May_2018.csv", index=False)
```

```python
import pandas as pd
df_june = pd.read_csv("Classes June 2018.csv")
df_june.head()
```

```python
df_june.columns = [
    "ActivitySiteID", 
    "ActivityDescription", 
    "BookingEndDate", 
    "BookingStartTime", 
    "MaxBookees", 
    "NumberBooked", 
    "PriceINR"
]
df_june.head()
```

```python
df_june["BookingEndDate"] = pd.to_datetime(df_june["BookingEndDate"], errors='coerce')
df_june["BookingStartTime"] = pd.to_datetime(df_june["BookingStartTime"], format="%H:%M:%S", errors='coerce').dt.time
df_june.head()
```

```python
df_june["DayOfWeek"] = df_june["BookingEndDate"].dt.day_name()
df_june["OccupancyRate"] = df_june["NumberBooked"] / df_june["MaxBookees"]
df_june.head()
```

```python
df_june.to_csv("Classes_June_2018.csv", index=False)
```

```python
df_combined = pd.concat([df_april_may, df_june], ignore_index=True)
print("Combined Data Shape:", df_combined.shape)
df_combined.head()
```

```python
print("Missing values after merging:")
print(df_combined.isnull().sum())
df_combined["PriceINR"].fillna(df_combined["PriceINR"].median(), inplace=True)
```

```python
duplicates = df_combined.duplicated().sum()
print("Duplicate records:", duplicates)
```

```python
invalid_occupancy = df_combined[df_combined["OccupancyRate"] > 1]
print("Invalid Occupancy Rate Records:", len(invalid_occupancy))
```

```python
df_combined["BookingEndDate"] = pd.to_datetime(df_combined["BookingEndDate"], errors='coerce')
quality_report = {
    "Total Records": len(df_combined),
    "Missing Values": df_combined.isnull().sum().to_dict(),
    "Duplicate Records": duplicates,
    "Invalid Occupancy Rate (>1)": len(invalid_occupancy),
    "Unique Activities": df_combined["ActivityDescription"].nunique(),
    "Date Range": (df_combined["BookingEndDate"].min(), df_combined["BookingEndDate"].max())
}
import pprint
pprint.pprint(quality_report)
```

```python
df_combined.to_csv("merging.csv", index=False)
```
